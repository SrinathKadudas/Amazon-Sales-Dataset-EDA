by looking at the data all the columns are string type, so we need to convert 'discounted_price',
       'actual_price', 'discount_percentage', 'rating', 'rating_count' as float type

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹','') # replace ₹ with empty
df['discounted_price'] = df['discounted_price'].str.replace(',','') # replace , with empty
df['discounted_price'] = df['discounted_price'].astype(float) # change datatype from string to float
```

```python
df['actual_price'] = df['actual_price'].str.replace('₹','') # replace ₹ with empty
df['actual_price'] = df['actual_price'].str.replace(',','') # replace , with empty
df['actual_price'] = df['actual_price'].astype(float) # change datatype from string to float
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%','') # replace % with empty
df['discount_percentage'] = df['discount_percentage'].astype(float) # change datatype from string to float
df['discount_percentage'] = df['discount_percentage']/100 
```

```python
df['rating'].value_counts()
```

```python
df.query('rating == "|"') # check the strange row
```

```python
df['rating'] = df['rating'].replace('|', '3.9') # product rating on Amazon by searching the provided product_id on their official website (amazon.in) The rating is 3.9. So, I am going to give the item rating a 3.9 as well.
```

```python
df['rating'] = df['rating'].astype(float)
```

```python
df['rating_count'] = df['rating_count'].str.replace(',','') #replace , with empty
df['rating_count'] = df['rating_count'].astype(float)
```

Dealing with the missing values

Dealing with the missing values is one of the most important part of the data wrangling process, we must deal with the missing values in order to get the correct insights from the data.
Missing Values

```python
df.isnull().sum().sort_values(ascending = False)
```

```python
df.isnull().sum().sum() # total number of missing values
```

```python
df['rating_count'].isnull().head()
```

```python
df[df['rating_count'].isnull()].head(5)
```

```python
df['rating_count'] = df.rating_count.fillna(value= df['rating_count'].median()) # replace the missing values with median of rating_count
```

```python
df.isnull().sum() # there are no null values now
```

```python
df.duplicated().any() # no duplicates in the dataset
```

```python
from sklearn.preprocessing import LabelEncoder
# label encode categorical variables

le_product_id = LabelEncoder()
le_category = LabelEncoder()
le_review_id = LabelEncoder()
le_review_content = LabelEncoder()
le_product_name = LabelEncoder()
le_user_name = LabelEncoder()
le_about_product = LabelEncoder()
le_user_id = LabelEncoder()
le_review_title = LabelEncoder()
le_img_link = LabelEncoder()
le_product_link = LabelEncoder()


df['product_id'] = le_product_id.fit_transform(df['product_id'])
df['category'] = le_category.fit_transform(df['category'])
df['review_id'] = le_review_id.fit_transform(df['review_id'])
df['review_content'] = le_review_content.fit_transform(df['review_content'])
df['product_name'] = le_product_name.fit_transform(df['product_name'])
df['user_name'] = le_user_name.fit_transform(df['user_name'])
df['about_product'] = le_about_product.fit_transform(df['about_product'])
df['user_id'] = le_user_id.fit_transform(df['user_id'])
df['review_title'] = le_review_title.fit_transform(df['review_title'])
df['img_link'] = le_img_link.fit_transform(df['img_link'])
df['product_link'] = le_product_link.fit_transform(df['product_link'])
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
# Calculate Pearson correlation coefficients (default in Pandas)
correlation_matrix = df.corr()
print(correlation_matrix)
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()
```
